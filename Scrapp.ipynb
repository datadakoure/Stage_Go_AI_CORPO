```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def url_to_soup(url:str):
    
    response = requests.get(url)
    
    # Vérifier que la requête a été réussie
    if response.status_code != 200:
        print(f"Erreur de requête: {response.status_code}")
        return []
    
    # Parser le contenu HTML de la page avec BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup
```

```python

def scrape_page(url):
    dic = {}
    dic2 = {}
    dic["mos"] = []
    dic["fr"] = []
    dic["en"] = []
    
    soup = url_to_soup(url)

    # Trouver les éléments contenant les phrases en mooré
    examples_contents = soup.find_all('span', class_="examplescontents")
    
    for examples_content in examples_contents:
        
        #Mooré
        phrases_moore_exemples = examples_content.find('span', class_="example").text.strip()

        try:
            

            phrases_others_exemples = examples_content.find('span', class_="translation")
            
    
            
            #Français
            phrases_fr_exemples = phrases_others_exemples.find("span", lang="fr").text.strip()
            
            #Englais
            phrases_en_exemples = phrases_others_exemples.find("span", lang="en").text.strip()
            dic["mos"].append(phrases_moore_exemples)
            dic["fr"].append(phrases_fr_exemples)
            dic["en"].append(phrases_en_exemples)
            
        except:
            ...
            # dic["mos"].append(phrases_moore_exemples)
            # dic["fr"].append(phrases_moore_exemples)
            # dic["en"].append(phrases_moore_exemples)

    
    #phrases = [phrase_moore.text.strip() for phrase_moore in phrases_moore_exemples ]
    
    #phrases_moore = soup.find_all('span', lang="mos")  # Remplacez 'vernacular' par la classe appropriée si nécessaire
    # Extraire les phrases en mooré
    return dic
```

```python
base_url = 'https://www.webonary.org/moore/browse/browse-vernacular-english/?key=mos&letter=a&lang=en'
soup= url_to_soup(base_url)
letters = soup.find_all('div', class_="lpTitleLetterCell")
letters = [letter.text.strip() for letter in letters]
```

```python

def extract_url(letters) -> list:
    urls = []
    for letter in letters:
        url_page = f"https://www.webonary.org/moore/browse/browse-vernacular-english/?key=mos&letter={letter}&lang=en"
        nbr_pages = url_to_soup(url_page).find("li", class_="page_info")
        
        if nbr_pages:
            
            nbr_pages=nbr_pages.text.split(" ")[-1]
            
            for i in range(1, int(nbr_pages)+1):
                url_page_2 = f"https://www.webonary.org/moore/browse/browse-vernacular-english/?key=mos&letter={letter}&lang=en&pagenr={i}"
                urls.append(url_page_2)
        else:
            urls.append(url_page)
    return urls

urls = extract_url(letters)
```

```python
def scrape_all_pages(base_url):
    
    dic = {}
    dic["mos"] = []
    dic["fr"] = []
    dic["en"] = []
   
    for url in base_url:
        print(url)
        
        url_dic = scrape_page(url)
        dic["mos"].extend(url_dic["mos"])
        dic["fr"].extend(url_dic["fr"])
        dic["en"].extend(url_dic["en"])
        
    return dic
```

```python
def save_to_csv(phrases, filename):
    df = pd.DataFrame(phrases)
    df.to_csv(filename,sep="\t", index=False)
    print(f"Données enregistrées dans {filename}")
```

```python
filename = 'phrases_moore2.csv'

# Scraper toutes les pages et sauvegarder les données dans un fichier CSV
phrases = scrape_all_pages(urls)
```
